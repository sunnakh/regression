### Residuals (Errors)

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(9, 6), dpi=80)

plt.scatter(df['Size(m2)'], df['Price'])

a = df['Size(m2)']
b = df['Price'] + (np.nanmean(df.Price) -  df['Price']) / 2
c = abs(np.nanmean(df.Price) -  df['Price']) / 2
s = np.nanmean(df.Price) -  df['Price']

plt.errorbar(a, b, yerr=c, fmt="none", ecolor = 'red')
for i in range(len(a)):
    plt.text(a[i], b[i], s[i], fontsize = 12)
plt.ylim(90000, 220000)
plt.axhline(y=np.nanmean(df.Price), color='black')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

p1, p2 = [60, 130], [120000, 170000] 
f = interp1d(p1, p2)
distances = [f(x) for x in df['Size(m2)']]


plt.figure(figsize=(9, 6), dpi=80)

plt.scatter(df['Size(m2)'], df['Price'])

a = df['Size(m2)']
b = df['Price'] + (distances -  df['Price']) / 2
c = abs(distances - df['Price']) / 2
s = (distances -  df['Price']).astype(int)

plt.errorbar(a, b, yerr=c, fmt="none", ecolor = 'red')
for i in range(len(a)):
    plt.text(a[i], b[i], s[i], fontsize = 12)
plt.ylim(0, 220000)
plt.plot([60, 130], [120000, 170000], color='black')
```

## Simple Linear Regression

We will now fit a simple line, the most simple that can be <br>
We will have a regression of _charges_ according to _age_

```python
import statsmodels.api as sm
```

```python
sns.heatmap(diamonds.corr(numeric_only=True), annot=True, vmin=-1, vmax=1, cmap="coolwarm")
```

```python
features = ["carat", "x"]

X = sm.add_constant(diamonds[features])
#X = sm.add_constant(diamonds[features])[0:10]
y = diamonds["price"]
#y = diamonds["price"][0:10]

model = sm.OLS(y, X).fit()

model.summary()
```

* [why the constant term ?](https://stackoverflow.com/questions/41404817/statsmodels-add-constant-for-ols-intercept-what-is-this-actually-doing)


t values help us to understand the prediction power of coefficients <br>
not very informative in simple linear regression, but will be more important when we add more variables <br>
const = intercept

```python
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as rmse
from sklearn.metrics import r2_score as r2

print(f"MAE: {mae(y, model.predict())}")
print(f"RMSE: {rmse(y, model.predict(), squared=False)}")
print(f"R2: {r2(y, model.predict())}")
```

### Train test split

```python
from sklearn.model_selection import train_test_split

features = ["carat", "x", "y", "z"]

X = sm.add_constant(diamonds[features])
y = diamonds["price"]

X, X_test, y, y_test = train_test_split(X, y, test_size=.2, random_state=2023)

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=.25, random_state=2024)
```

### Cross Validation

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as rmse
from sklearn.metrics import r2_score as r2


cv = KFold(n_splits=5, shuffle=True, random_state=2023)

# Create a list to store validation scores for each fold
r2_list = []
mae_list = []
rmse_list = []

# Loop through each fold in X and y
for train_ind, val_ind in cv.split(X, y):
    # Subset data based on CV folds
    X_train, y_train = X.iloc[train_ind], y.iloc[train_ind]
    X_val, y_val = X.iloc[val_ind], y.iloc[val_ind]
    # Fit the Model on fold's training data
    model = sm.OLS(y_train, X_train).fit()
    # Append Validation score to list 
    r2_list.append(r2(y_val, model.predict(X_val)))
    mae_list.append(mae(y_val, model.predict(X_val)))
    rmse_list.append(rmse(y_val, model.predict(X_val), squared = False))

print("All Validation R2s: ", [round(x, 3) for x in r2_list])
print(f"Cross Val R2s: {round(np.mean(r2_list), 3)}")

print("All Validation MAEs: ", [round(x, 3) for x in mae_list])
print(f"Cross Val MAEs: {round(np.mean(mae_list), 3)}")

print("All Validation RMSEs: ", [round(x, 3) for x in rmse_list])
print(f"Cross Val RMSEs: {round(np.mean(rmse_list), 3)}")
```

## Regularization

```python
features = [col for col in diamonds.columns if col != 'price']

X = sm.add_constant(diamonds[features])
#X = sm.add_constant(diamonds[features])[0:10]
y = diamonds["price"]

X, X_test, y, y_test = train_test_split(X, y, test_size=.2, random_state=2023)
```

```python
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_absolute_error as mae

n_alphas = 200
alphas = 10 ** np.linspace(-3, 3, n_alphas)

ridge_model = RidgeCV(alphas=alphas, cv=5)

ridge_model.fit(X, y)
print(ridge_model.score(X, y))
print(mae(y, ridge_model.predict(X)))
print(ridge_model.alpha_)
```

```python
ridge_model.score(X_test, y_test)
```

```python
from sklearn.linear_model import LassoCV
from sklearn.metrics import mean_absolute_error as mae

n_alphas = 200
alphas = 10 ** np.linspace(-3, 3, n_alphas)

lasso_model = LassoCV(alphas=alphas, cv=5)

lasso_model.fit(X, y)
print(lasso_model.score(X, y))
print(mae(y, lasso_model.predict(X)))
print(lasso_model.alpha_)
```

```python
lasso_model.score(X_test, y_test)
```

```python
from sklearn.linear_model import ElasticNetCV

alphas = 10 ** np.linspace(-3, 3, 200)
l1_ratios = np.linspace(.9, 1, 10) # .9, 1, 10

enet_model = ElasticNetCV(alphas=alphas, l1_ratio=l1_ratios, cv=5)

enet_model.fit(X, y)
print(enet_model.score(X, y))
print(enet_model.score(X_test, y_test))
print(enet_model.alpha_)
print(enet_model.l1_ratio_)
```
